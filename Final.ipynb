```python
dosya = "/content/drive/MyDrive/Colab Notebooks/Veri.txt"


def menu():
    print("Katalog Kayıtları\n")
    print("1. Katalog Kayıtlarını Listele")
    print("2. Kataloğa Yeni Kayıt Ekle")
    print("3. Kataloğu Sil")
    print("4. Kataloğu Güncelle")
    print("5. Kataloğu Temizle")
    print("6. Katalogda Kayıt Ara")
    print()

    sec = input("Lütfen yapmak istediğiniz işlemi seçiniz (1-2-3-4-5-6-0): ")
    if sec == "0":
        print("Katalogtan Çıkış Yaptınız...!")
    elif sec == "1":
        listele()
    elif sec == "2":
        ekle()
    elif sec == "3":
        sil()
    elif sec == "4":
        güncelle()
    elif sec == "5":
        temizle()
    elif sec == "6":
        ara()
    else:
        print("Lütfen bu işlemlerden birini seçin (1/2/3/4/5/6/0)")
        menu()


def listele():
    islem = open(dosya)
    for satir in islem:
        print(satir.strip())
    islem.close()
    menu()


def ekle():
    islem = open(dosya, "a", encoding="utf-8")
    eserAdı = input("Eserin adını giriniz: ")
    yazar = input("Yazarın adını giriniz: ")
    basim = input("Basım yılını giriniz: ")
    islem.write(eserAdı.title() + "-" + yazar.title() + "-" + basim + "\n")
    islem.close()
    menu()


def güncelle():
    islem = open(dosya, "r", encoding="utf-8")
    kayitlar = islem.readlines()
    islem.close()

    print("Mevcut kayıtlar:")
    for satir in kayitlar:
        eser = satir.strip()
        ayir = eser.split("-")
        eserAdı = ayir[0]
        print(eserAdı)
    güncelleme = input("Lütfen güncellenecek eserin adını giriniz: ")

    islem = open(dosya, "w", encoding="utf-8")
    for satir in kayitlar:
        eser = satir.strip()
        ayir = eser.split("-")
        eserAdı = ayir[0]
        if eserAdı == güncelleme.title():
            yeniAd = input("Yeni eser adını giriniz: ")
            yeniYazar = input("Yeni yazar adını giriniz: ")
            yeniBasim = input("Yeni basım yılını giriniz: ")
            kayit = yeniAd.title() + "-" + yeniYazar.title() + "-" + yeniBasim
            islem.write(kayit + "\n")
        else:
            islem.write(satir)
    islem.close()
    menu()


def ara():
    islem = open(dosya, "r", encoding="utf-8")
    arama = input("Aranan eserin adını giriniz: ")
    bulundu = False
    for satir in islem:
        eser = satir.strip()
        ayir = eser.split("-")
        eserAdı = ayir[0]
        if eserAdı == arama.title():
            print("Arama sonucu: " + satir)
            bulundu = True
    islem.close()
    if not bulundu:
        print("Aranan eser bulunamadı.")
    menu()


def temizle():
    onay = input("Kataloğu temizlemek istediğinizden emin misiniz? (E/H): ")
    if onay.lower() == "e":
        islem = open(dosya, "w")
        islem.close()
        print("Katalog temizlendi.")
    menu()


def sil():
    islem = open(dosya, "r", encoding="utf-8")
    kayitlar = islem.readlines()
    islem.close()

    silinecek = input("Silinecek eserin adını giriniz: ").title()

    islem = open(dosya, "w", encoding="utf-8")
    silindi = False
    for satir in kayitlar:
        eser = satir.strip()
        ayir = eser.split("-")
        eserAdı = ayir[0]
        if eserAdı != silinecek:
            islem.write(satir)
        else:
            silindi = True
    islem.close()

    if silindi:
        print("Eser başarıyla silindi.")
    else:
        print("Silinecek eser bulunamadı.")
    menu()


menu()
  
```
